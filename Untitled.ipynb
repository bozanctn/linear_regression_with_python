```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
#data = pd.read_excel("Data Dictionary - carprices.xlsx")
df = pd.read_csv("CarPrice_Assignment.csv")
```

```python
df
```

```python
df[["enginesize", "price"]].corr()
```

```python
X = df[["enginesize"]]
y = df[["price"]]
reg_model = LinearRegression().fit(X, y)
```

```python
# constant - (bias)
reg_model.intercept_[0]
```

```python
#coefficent of engine size
reg_model.coef_[0][0]
```

```python
g = sns.regplot(x=X, y=y, scatter_kws={"color" : "b", "s" : 9},
                ci=False, color="r")
g.set_title(f"Model Equation: Price = {round(reg_model.intercept_[0], 2)} + Car Engine Power * {round(reg_model.coef_[0][0], 2)}")
g.set_ylabel("Car Price")
g.set_xlabel("Engine Power of Car")
plt.xlim(40, 300)
plt.ylim(bottom=0)
plt.show
```

```python
y_pred = reg_model.predict(X)
mean_absolute_error(y, y_pred)
```
